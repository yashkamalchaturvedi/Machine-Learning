```python
import numpy as np
import matplotlib.pyplot as plt

# pyplot is mainly intended for interactive plots and simple cases of programmatic plot generation:
plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x=[1,2,3]
y=[3,2,1]
a=[11,22,33]
b=[33,22,11]

plt.scatter(x,y, label="first", linewidth=10)
plt.plot(a,b, label="second")
plt.legend()
plt.show()
```
